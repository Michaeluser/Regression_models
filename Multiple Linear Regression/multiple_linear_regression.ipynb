```python
import pandas as pd
import numpy as np
import matplotlib as mt
import random
```

```python
dataframe = pd.read_csv("ds_salaries.csv")
dataframe.describe()
```

```python
dataframe.head()
dataframe.experience_level.unique()
dataframe.company_size.unique()
dataframe.remote_ratio.unique()
```

```python

for i in list(dataframe.index):
    if dataframe.loc[i, "experience_level"] == "MI":
        dataframe.loc[i, "experience_level"] = 10
    if dataframe.loc[i, "experience_level"] == "SE":
        dataframe.loc[i, "experience_level"] = 20
    if dataframe.loc[i, "experience_level"] == "EN":
        dataframe.loc[i, "experience_level"] = 30
    if dataframe.loc[i, "experience_level"] == "EX":
        dataframe.loc[i, "experience_level"] = 40
dataframe.head()
```

```python
 for i in list(dataframe.index):
        if dataframe.loc[i, "company_size"] == "S":
            dataframe.loc[i, "company_size"] = 15
        if dataframe.loc[i, "company_size"] == "M":
            dataframe.loc[i, "company_size"] = 25
        if dataframe.loc[i, "company_size"] == "L":
            dataframe.loc[i, "company_size"] = 35
            
dataframe.head()
```

```python
print(list(dataframe.columns))
for col in dataframe.columns.values:
    if col != "experience_level" and col != "salary_in_usd" and col != "company_size" and col != "remote_ratio":
        del dataframe[col]
dataframe["B"] = [1 for i in range(len(dataframe))]
dataframe
```

```python
y_train = [dataframe.loc[i, "salary_in_usd"] for i in range(len(dataframe))]
x_train = dataframe[["B", "experience_level", "remote_ratio", "company_size"]].to_numpy()
```

```python
print(x_train)
th = np.dot(x_train.T, x_train)
b_slopes = np.dot(np.dot(np.linalg.inv((th).astype("float64")), x_train.T), y_train)
b_slopes = b_slopes.astype("float64")
print(b_slopes)
slopes_sorted = {"B":b_slopes[0], "experience_level":b_slopes[1], "remote_ratio":b_slopes[2], "company_size":b_slopes[3]}
slopes_sorted["company_size"]
```

```python
test_dataf = pd.DataFrame({
"B":[1 for i in range(1000)], 
"experience_level":[random.choice([30, 10, 20, 40]) for i in range(1000)], 
"remote_ratio":[random.choice([50, 0, 100]) for i in range(1000)], 
"company_size":[random.choice([25, 15, 35]) for i in range(1000)],
})
test_dataf

test_dataf["salary_in_usd"] = [(slopes_sorted["B"] + slopes_sorted["experience_level"]*test_dataf.iloc[i].experience_level + slopes_sorted["remote_ratio"]*test_dataf.iloc[i].remote_ratio + slopes_sorted["company_size"]*test_dataf.iloc[i].company_size) for i in range(1000)]
sorted(list(test_dataf.salary_in_usd))[-1]
```
